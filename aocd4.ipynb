```python
#partone
start = open('/Users/oskarsoderbom/aocd4.txt').read()

test1 = start.replace('\n', ',')
testar = test1.split(',')
startlist = []
for line in testar:

    startlist.append(list(line))

exlist = [['2', '4'],['6', '8'],
['2', '3'],['4','5'],
['5', '7'],['7', '9'],
['2', '8'],['3', '7'],
['6', '6'],['4', '6'],
['2', '6'],['4', '8']]



biggestlist = []
for i in range(len(startlist)):
    for j in range(len(startlist[i])):
        if startlist[i][j] == '-':
            x = startlist[i][0:j]
            y = startlist[i][j+1:len(startlist)+1]
            xstr = ''.join(map(str, x))
            ystr = ''.join(map(str, y))

            biggestlist.append([xstr,ystr])

counter = 0
for lista in range(0, len(biggestlist), 2):
    if (int(biggestlist[lista][0]) >= int(biggestlist[lista+1][0])) and (int(biggestlist[lista][1]) <= int(biggestlist[lista+1][1])):
            counter += 1
    elif (int(biggestlist[lista][0]) <= int(biggestlist[lista+1][0])) and (int(biggestlist[lista][1]) >= int(biggestlist[lista+1][1])):
            counter +=1
            print(biggestlist[lista], biggestlist[lista+1])
print(counter)
```
