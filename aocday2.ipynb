```python
stratlist = list(strat.replace('\n', ','))
exlist = ['A', ' ', 'Y','B',' ','X', 'C', ' ', 'Z']
stratlist

new_list = []
for i in stratlist:
    if i ==',':
        pass
    else:
        new_list.append(i)
new_list
```

```python
points = {'A': 1, 'B': 2, 'C':3, 'X':1, 'Y': 2, 'Z': 3}
```

```python
score = 0
for count,value in enumerate(new_list):

    if value == ' ':  #hittar mittpunkten på varje lista och extraherar värden till höger och vänster
        value1 = points.get(new_list[count-1])
        value2 = points.get(new_list[count+1])
        #print(value1, value2)

        if value1 == value2:      #om värdena är desamma så är det allitd lika
            score += value2 + 3
        
        elif ((value1 == 1 and value2 ==3) #alla förluster
        or (value1 == 2 and value2 == 1) 
        or (value1== 3 and value2 == 2)):
            score += value2
        
        else:
            score += value2 + 6
        #print(score)
print(score)
```
